```python
import matplotlib.pyplot as plt

# Data
file_sizes = [200, 400, 600, 800, 1000]  # File sizes in megabytes
time_python = [1.097, 2.622, 3.803, 5.071, 7.833]  # Time taken in seconds by Python
time_cpp = [20.906, 40.145, 60.668, 77.39, 103.278]     # Time taken in seconds by C++
time_java = [3.979, 5.734, 6.948, 8.93, 13.778]    # Time taken in seconds by Java

# Create the plot
plt.figure()

# Plot for Python
plt.plot(file_sizes, time_python, marker='o', label='Python', color='blue')

# Plot for C++
plt.plot(file_sizes, time_cpp, marker='s', label='C++', color='red')

# Plot for Java
plt.plot(file_sizes, time_java, marker='^', label='Java', color='green')

# Adding titles and labels
plt.title('Performance Comparison')
plt.xlabel('File Size (MB)')
plt.ylabel('Time Taken (seconds)')

# Adding a legend
plt.legend()

# Display the plot
plt.grid(True)
plt.show()
plt.close()
```
